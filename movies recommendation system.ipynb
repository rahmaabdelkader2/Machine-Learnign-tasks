The function creates a pivot table from the combined_movies_data DataFrame, where the rows represent user IDs, the columns represent movie titles, and the values represent the corresponding ratings given by users to movies.
The function prints the size of the pivot table (number of rows and columns) and returns the pivot table.

```python
def movies_rating_crosstable(combined_movies_data):
    movies_rating_crosstable_df = pd.pivot_table(
        data=combined_movies_data,
        values='rating',
        index='userId',
        columns='title',
        fill_value=0
    )
    print("The Pivot_table size = (userIDs, movie titles) = ", movies_rating_crosstable_df.shape)
    return movies_rating_crosstable_df
```

```python
# creating the (Collaborative Filtering) CF tabel
movies_rating_transposed=movies_rating_crosstable(combined_movies_data)
```

### Transposing it to calculate the similarity matrix of item to item to be able to get the similarity scores

```python
user_ratings_to_movies = movies_rating_transposed.transpose()
print("transpose Pivot tabel size = (movie titles , userIDs) = ",user_ratings_to_movies.shape)
```

```python
user_ratings_to_movies.head(10)
```

### Calculate the cosine similarity between items (movies)

```python
from sklearn.metrics.pairwise import cosine_similarity

def calculate_item_similarity(user_ratings_to_movies):
    
    similarity_scores_movies = cosine_similarity(user_ratings_to_movies)
    print("Shape of the similarity matrix: ", similarity_scores_movies.shape)
    print(similarity_scores_movies)
    return similarity_scores_movies
```

```python
similarity_scores_movies=calculate_item_similarity(user_ratings_to_movies)
```

### Creating dataframe with the movies titles and similarity

```python
data_matrix_item = pd.DataFrame(data=similarity_scores_movies, index=user_ratings_to_movies.transpose().columns, columns=user_ratings_to_movies.transpose().columns)
```

```python
data_matrix_item.head(10)
```

### function that takes the name of the movie and the number of recommendations

```python
def recommend_movies_similar_to_movie(movie, recomemnds=10):
    movie = movie
    number_of_recommendation = recomemnds

    # in data_matrix movie titles were set to be the indices ad then saving the indices means savng the mavie titles
    indices = data_matrix_item.loc[movie].nlargest(
        number_of_recommendation+1).index

    print("The recommended movies if u like '{}' are \n".format(movie))
    for idx, num in enumerate(indices):
        if idx > 0:
            print(num)
```

### Recommend movies for a certain user based on similar users

```python
def recommend_movies_for_user(userID, recommends, similar_users=5):
    #catching function args
    userID = userID
    number_of_similar_users = similar_users
    number_of_recommendation = recommends

    counter = 0
    # save the ids of similar users
    indices = data_matrix_user.loc[userID].nlargest(number_of_similar_users+1).index

    print("the recommended movies for user {}".format(userID))

    # loop on similar users and print indices
    
    for idx, num in enumerate(indices):
        # save all the movies that this user watched with high ratings
        movie_rec_user = movies_rating_transposed.loc[indices[idx]].nlargest(
            number_of_recommendation+1).index

        # lopp over those movies with high ratings
        for idx, num in enumerate(movie_rec_user):

            # check if the user have watched it or not (if user rated that movie or not)
            if (movies_rating_transposed[movie_rec_user[idx]].loc[userID] == 0) & (counter < number_of_recommendation):

                # if he haven't watched it recommend the movie
                print(movie_rec_user[idx])
                counter += 1
```
